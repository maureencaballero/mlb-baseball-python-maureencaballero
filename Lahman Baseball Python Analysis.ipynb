## Fielding Table Groupings and Putouts

###Using the fielding table, group players into three groups based on their position: label ###players with position OF as "Outfield", those with position "SS", "1B", "2B", and "3B" as ###"Infield", and those with position "P" or "C" as "Battery". Determine the number of putouts ###made by each of these three groups in 2016.

```python
fieldingdf = pd.read_sql('SELECT * FROM fielding;', con=engine)
fieldingdf.head()
```

```python
fieldingdf['group'] = fieldingdf['pos']
fieldingdf.head()
```

```python
#df['column name'] = df['column name'].replace(['old value'],'new value')
fieldingdf['group'] = fieldingdf['group'].replace(['SS'],'Infield')
fieldingdf['group'] = fieldingdf['group'].replace(['1B'],'Infield')
fieldingdf['group'] = fieldingdf['group'].replace(['2B'],'Infield')
fieldingdf['group'] = fieldingdf['group'].replace(['3B'],'Infield')
fieldingdf['group'] = fieldingdf['group'].replace(['OF'],'Outfield')
fieldingdf['group'] = fieldingdf['group'].replace(['P'],'Battery')
fieldingdf['group'] = fieldingdf['group'].replace(['C'],'Battery')
fieldingdf.head()
```

# 3. WORLD SERIES QUESTIONS

###a. From 1970 â€“ 2016, what is the largest number of wins for a team that did not win the world series?  SEATTLE 116

```python
teamsdf = pd.read_sql('SELECT * FROM teams;', con=engine)
teamsdf.head()

```

```python
teams19702016 = teamsdf[(teamsdf['yearid'] >= 1970) & (teamsdf['yearid'] <= 2016)]
teams19702016
teams19702016.info()
```

```python
teams19702016.head()
```

```python
maxteams19702016 = teams19702016[teams19702016['wswin'] == 'N']
maxwins = maxteams19702016.groupby('teamid')['w'].max()
maxwins
```

```python
maxwins = maxwins.sort_values(ascending = False)
maxwins
```

###b. What is the smallest number of wins for a team that did win the world series?  63, LAN in 1981

```python
minteams19702016 = teams19702016[teams19702016['wswin'] == 'Y']
minwins = minteams19702016.groupby(['teamid', 'yearid'])['w'].min()
minwins = minwins.sort_values()
minwins
```

```python
someteams19702016 = teams19702016[(teams19702016['wswin'] == 'Y') & (teams19702016['yearid'] != 1981)]
someminwins = someteams19702016.groupby(['teamid','yearid'])['w'].min()
someminwins = someminwins.sort_values()
someminwins
```
